```python
#Creo un diccionario donde cada llave son los equipos
gol = {}
games_played = list(archivo.columns.values)
```

```python
for game in games_played:
  local = archivo[game]['matchCentreData']['home']['name']
  if local not in gol.keys():
    gol[local]= {'goles':0,'partidos':0,'ganados':0,'perdidos':0,'empatados':0}
  visitante = archivo[game]['matchCentreData']['away']['name']
  if visitante not in gol.keys():
    gol[visitante]= {'goles':0,'partidos':0,'ganados':0,'perdidos':0,'empatados':0}
```

```python
#Sumo los goles en cada equipo
for game in games_played:
  df2 = archivo[game]['matchCentreData']
  local = df2['home']['name']
  gollocal = df2['score'][0]
  visitante = df2['away']['name']
  golvisita = df2['score'][4]
  # Sumo los goles en el diccionario
  gol[local]['goles'] = gol[local]['goles'] + int(gollocal)
  gol[visitante]['goles']=gol[visitante]['goles'] + int(golvisita)
  #Sumo los partidos en el diccionario
  gol[local]['partidos'] = gol[local]['partidos'] + 1
  gol[visitante]['partidos']=gol[visitante]['partidos'] + 1
  # Sumo las victorias, empates y perdidas
  if gollocal > golvisita:
   gol[local]['ganados'] = gol[local]['ganados'] + 1
   gol[visitante]['perdidos']=gol[visitante]['perdidos'] + 1
  elif gollocal < golvisita:
   gol[local]['perdidos'] = gol[local]['perdidos'] + 1
   gol[visitante]['ganados']=gol[visitante]['ganados'] + 1
  else:
   gol[local]['empatados'] = gol[local]['empatados'] + 1
   gol[visitante]['empatados']=gol[visitante]['empatados'] + 1      
```

```python
years = ["2016", "2017", "2018", "2019", "2020", "2021"]
```

```python
games_played = {}
def gamesPlayed(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                match_played = player['stats']
                if player_id not in result:
                    result[player_id] = 0
                if match_played != {}:
                    result[player_id] += 1
    return result

for year in years:
    games_played[f"games_{year}"] = gamesPlayed(year)
```

```python
MVP = {}
def best_players(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                match_mvp = player['isManOfTheMatch']
                if player_id not in result:
                    result[player_id] = 0
                if match_mvp is True:
                    result[player_id] += 1
    return result

for year in years:
    MVP[f"mvp_{year}"] = best_players(year)
```

```python
gkSaves = {}
def gk_saves(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    total_saves=player['stats']['totalSaves']
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position == 'GK':
                        for save in total_saves:    
                            result[player_id] += 1
                except KeyError:
                    1
    return result

for year in years:
    gkSaves[f"saves_{year}"] = gk_saves(year)
```

```python
passSuccess = {}
def success_passes(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    passessuccess = player['stats']['passSuccess'].values()
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position not in ('GK'):
                        passes = sum(passessuccess)/(len(passessuccess)*100)
                        result[player_id] += passes
                except KeyError:
                    1
    return result

for year in years:
    passSuccess[f"passesSuccess_{year}"] = success_passes(year)
```

```python
tackleSuccess = {}
def tackle_Success(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    tacklesuccess = player['stats']['tackleSuccess'].values()
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position not in ('GK'):
                        tackles = sum(tacklesuccess)/(len(tacklesuccess)*100)
                        result[player_id] += tackles
                except KeyError:
                    1
    return result

for year in years:
    tackleSuccess[f"successTackles_{year}"] = tackle_Success(year)
```

```python
dribbleSuccess = {}
def drible_Success(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    driblesuccess = player['stats']['dribbleSuccess'].values()
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position not in ('GK'):
                        dribles = sum(driblesuccess)/(len(driblesuccess)*100)
                        result[player_id] += dribles
                except KeyError:
                    1
    return result

for year in years:
    dribbleSuccess[f"successDribles_{year}"] = drible_Success(year)
```

```python
Interception = {}
def interceptions_success(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    interception = player['stats']['interceptions'].values()
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position not in ('GK'):
                        interceptions = sum(interception)
                        result[player_id] += interceptions
                except KeyError:
                    1
    return result

for year in years:
    Interception[f"interceptions_{year}"] = interceptions_success(year)
```

```python
foulsCommited = {}
def fouls_Commited(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            players = tournament[game]['matchCentreData'][where]['players']
            for player in players:
                player_id = player['playerId']
                player_position = player['position']
                try:
                    foul = player['stats']['foulsCommited'].values()
                    if player_id not in result:
                        result[player_id] = 0
                    if player_position not in ('GK'):
                        fouls = sum(foul)
                        result[player_id] += fouls
                except KeyError:
                    1
    return result

for year in years:
    foulsCommited[f"foulsCommited_{year}"] = fouls_Commited(year)
```

### Se arma tabla donde se agrupan los eventos relevantes de cada partido por jugador

```python
events_goal = {}
def goal_players(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            events=tournament[game]['matchCentreData'][where]['incidentEvents']
            for event in events:
                try:
                    player_id = event['playerId']
                    goal = event['isGoal']
                    if player_id not in result:
                        result[player_id] = 0
                    if goal is True:
                        result[player_id] += 1
                except KeyError:
                    1
    return result

for year in years:
    events_goal[f"goal_{year}"] = goal_players(year)
```

```python
events_assist = {}
def assist_players(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            events=tournament[game]['matchCentreData'][where]['incidentEvents']
            for event in events:
                try:
                    player_id = event['relatedPlayerId']
                    goal = event['isGoal']
                    if player_id not in result:
                        result[player_id] = 0
                    if goal is True:
                        result[player_id] += 1
                except KeyError:
                    1
    return result

for year in years:
    events_assist[f"assist_{year}"] = assist_players(year)
```

```python
yellow_card = {}
def card_players(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            events=tournament[game]['matchCentreData'][where]['incidentEvents']
            for event in events:
                try:
                    player_id = event['playerId']
                    card = event['cardType']['displayName']
                    if player_id not in result:
                        result[player_id] = 0
                    if card in ("Yellow"):
                        result[player_id] += 1
                except KeyError:
                    1
    return result

for year in years:
    yellow_card[f"yellow_card_{year}"] = card_players(year)
```

```python
red_card = {}
def red_card_players(year):
    result = {}
    games = globals()[f"games_{year}"]
    tournament = globals()[f"torneo_{year}"]
    for game in games:
        for where in ('home','away'):
            events=tournament[game]['matchCentreData'][where]['incidentEvents']
            for event in events:
                try:
                    player_id = event['playerId']
                    card = event['cardType']['displayName']
                    if player_id not in result:
                        result[player_id] = 0
                    if card in ("SecondYellow"):
                        result[player_id] += 1
                except KeyError:
                    1
    return result

for year in years:
    red_card[f"red_card_{year}"] = red_card_players(year)
```
